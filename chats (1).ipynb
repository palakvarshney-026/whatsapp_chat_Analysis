```python
import re
import pandas as pd
```

```python
f= open("chatss.txt",'r', encoding= 'utf8')
```

```python
data= f.read()
```

```python
data
```

```python
pattern= '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
message= re.split(pattern, data)[1:]
```

```python
message
```

```python
date= re.findall(pattern, data)
```

```python
date
```

```python
df= pd.DataFrame({"user_date": date, "user_messagee": message})
```

```python
df
```

```python
df["user_date"]
```

```python
df["user_date"]= pd.to_datetime(df["user_date"], format= "%d/%m/%y, %H:%M - ") 
```

```python
users= []
messages= []
for message in df['user_messagee']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append("".join(entry[2:]))
    else:
        users.append('group notification')
        messages.append(entry[0])
df['user']= users
df['message']= messages
df.drop(columns= ['user_messagee'], inplace= True)
    
```
